### Pregunta 6: Procesar datos JSON (1 punto)

Escribe una funcion `analizar_inventario(json_texto)` que reciba un **string JSON** representando una lista de productos y retorne un **diccionario** con:

- `"total_productos"`: cantidad de productos en la lista
- `"valor_total"`: suma de `precio * cantidad` de todos los productos
- `"producto_mas_caro"`: nombre del producto con el precio mas alto

Debes usar `json.loads()` para parsear el string.

````python
# Pregunta 6: Analizar inventario desde JSON
import json

inventario_json = json.dumps([
    {"nombre": "Laptop", "precio": 1200000, "cantidad": 5},
    {"nombre": "Mouse", "precio": 35000, "cantidad": 20},
    {"nombre": "Teclado", "precio": 80000, "cantidad": 15},
    {"nombre": "Monitor", "precio": 450000, "cantidad": 8}
])

# Escribe tu funcion aqui
def analizar_inventario(json_texto):
    productos = json.loads(json_texto)
    total_productos = len(productos)
    valor_total = 0
    precio_mas_caro = ""
    precio_maximo = 0

    for p in productos:
        valor_total += p["precio"] * p["cantidad"]
        if p["precio"] > precio_maximo:
            precio_maximo = p["precio"]
            precio_mas_caro = p["nombre"]
    return {
        "total_productos": total_productos,
        "valor_total": valor_total,
        "producto_mas_caro": precio_mas_caro
    }
    pass  


# Usa tu funcion
resultado_inventario = analizar_inventario(inventario_json)

print("Resultado:", resultado_inventario)
````

### Pregunta 7: Redimensionar imagen proporcional (1 punto)

En Pillow, al redimensionar una imagen es importante mantener la proporcion (aspect ratio). Escribe una funcion `calcular_nuevo_tamano(ancho, alto, max_dim)` que:

1. Determine cual dimension (ancho o alto) es la mayor
2. Calcule el factor de escala para que la mayor dimension sea igual a `max_dim`
3. Aplique el mismo factor a ambas dimensiones
4. **Retorne** una tupla `(nuevo_ancho, nuevo_alto)` con valores **enteros** (usa `int()`)

Ejemplo: una imagen de 800x600 con `max_dim=400` debe retornar `(400, 300)`.

````python
# Pregunta 7: Calcular nuevo tamano manteniendo proporcion

# Escribe tu funcion aqui
def calcular_nuevo_tamano(ancho, alto, max_dim):

    if ancho > alto:
        proporcion = max_dim / ancho
    else:
        proporcion = max_dim / alto

    nuevo_ancho = ancho * proporcion
    nuevo_alto = alto * proporcion

    return int(nuevo_ancho), int(nuevo_alto)

pass 


# Pruebas
print("800x600, max 400:", calcular_nuevo_tamano(800, 600, 400))
print("600x800, max 400:", calcular_nuevo_tamano(600, 800, 400))
print("1024x768, max 512:", calcular_nuevo_tamano(1024, 768, 512))
````

### Pregunta 8: Preparar solicitud para API de Hugging Face (1 punto)

Escribe una funcion `preparar_solicitud(api_key, prompt, sistema=None, max_tokens=150, temperature=0.7)` que retorne un **diccionario** con dos claves:

- `"headers"`: diccionario con `"Authorization"` (formato `"Bearer {api_key}"`) y `"Content-Type"` (`"application/json"`)
- `"payload"`: diccionario con `"model"` (`"Qwen/Qwen2.5-7B-Instruct"`), `"messages"` (lista de mensajes), `"max_tokens"` y `"temperature"`

Si `sistema` no es `None`, la lista de mensajes debe tener primero un mensaje con `role: "system"` y luego el mensaje del usuario. Si `sistema` es `None`, solo el mensaje del usuario.

````python
# Pregunta 8: Preparar solicitud completa para HF API

# Escribe tu funcion aqui
def preparar_solicitud(api_key, prompt, sistema=None, max_tokens=150, temperature=0.7):

    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }

    mensaje = []

    if sistema is not None:
        mensaje.append({"role": "system", "content": sistema})
    mensaje.append({"role": "user", "content": prompt})
    payload = {
        "model": "Qwen/qwen2.5-7B-Instruct",
        "messages": mensaje,
        "max_tokens": max_tokens,
        "temperature": temperature
    }

    return {
        "headers": headers,
        "payload": payload
    }
    pass 


# Pruebas
print("=== Sin sistema ===")
r1 = preparar_solicitud("hf_abc123", "Hola")
print(json.dumps(r1, indent=2))

print("\n=== Con sistema ===")
r2 = preparar_solicitud("hf_abc123", "Hola", sistema="Eres un tutor.", max_tokens=100)
print(json.dumps(r2, indent=2))
````

### Pregunta 9: Construir prompt few-shot (1 punto)

La tecnica few-shot incluye ejemplos dentro del prompt. Escribe una funcion `construir_prompt_clasificacion(ejemplos, texto_nuevo)` que:

1. Reciba `ejemplos`: una lista de tuplas `(texto, sentimiento)`
2. Reciba `texto_nuevo`: el texto a clasificar
3. **Retorne** un string con el prompt completo usando este formato exacto:

```
Clasifica el sentimiento como: positivo, negativo o neutro.

Texto: "ejemplo1"
Sentimiento: sentimiento1

Texto: "ejemplo2"
Sentimiento: sentimiento2

Texto: "texto_nuevo"
Sentimiento:
```

````python
# Pregunta 9: Construir prompt few-shot para clasificacion

# Escribe tu funcion aqui
def construir_prompt_clasificacion(ejemplos, texto_nuevo):
    prompt = "Clasifica el siguiente texto como positivo, negativo o neutro:\n\n"
    for texto, sentimiento in ejemplos:
        prompt += f"Texto: {texto}\n"
        prompt += f"Sentimiento: {sentimiento}\n\n"
    prompt += f"Texto: {texto_nuevo}\n"
    prompt += "Sentimiento:"
    return prompt

    pass  


# Prueba
mis_ejemplos = [
    ("Me encanto la pelicula", "positivo"),
    ("Fue terrible", "negativo"),
    ("La reunion es a las 3", "neutro")
]

prompt_resultado = construir_prompt_clasificacion(mis_ejemplos, "El hotel fue fantastico")
print(prompt_resultado)
````

### Pregunta 10: Gestionar historial de chatbot (1 punto)

Escribe dos funciones para gestionar el historial de un chatbot:

**Funcion 1:** `agregar_turno(historial, mensaje_usuario, respuesta_asistente)` que:
- Agregue al `historial` un diccionario `{"role": "user", "content": mensaje_usuario}`
- Luego agregue un diccionario `{"role": "assistant", "content": respuesta_asistente}`
- **Retorne** el historial actualizado

**Funcion 2:** `preparar_mensajes(historial, sistema)` que:
- Cree una **nueva lista** que comience con `{"role": "system", "content": sistema}`
- Luego agregue **todos** los mensajes del historial
- **Retorne** la lista completa (system + historial)

````python
# Pregunta 10: Gestionar historial de chatbot

# Escribe tus funciones aqui
def agregar_turno(historial, mensaje_usuario, respuesta_asistente):
    historial.append({"role": "user", "content": mensaje_usuario})
    historial.append({"role": "assistant", "content": respuesta_asistente})
    return historial
    pass  

def preparar_mensajes(historial, sistema):
    mensajes_completos = [{"role": "system", "content": sistema}] 
    mensajes_completos = mensajes_completos + historial
    return mensajes_completos
    pass  


# Prueba
mi_historial = []
mi_historial = agregar_turno(mi_historial, "Hola", "Hola! En que te ayudo?")
mi_historial = agregar_turno(mi_historial, "Que es Python?", "Python es un lenguaje de programacion.")

print("Historial:")
for m in mi_historial:
    print(f"  [{m['role']}]: {m['content']}")

print("\nMensajes para API:")
mensajes_api = preparar_mensajes(mi_historial, "Eres un tutor.")
for m in mensajes_api:
    print(f"  [{m['role']}]: {m['content']}")
````
