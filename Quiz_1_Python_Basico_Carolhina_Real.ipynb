### Pregunta 7: Variables y operaciones (1 punto)

Una tienda ofrece un descuento del 25% en productos que cuestan mas de $100.

Dado un `precio_original` de `160`, calcula el `precio_final` aplicando el descuento.

Guarda el resultado en la variable `precio_final`.

````python
# Pregunta 7: Calcula el precio con descuento
precio_original = 160
porcentaje_descuento = 25

# Escribe tu codigo aqui
precio_final = precio_original - (precio_original * porcentaje_descuento / 100) 


print("Precio original: $", precio_original)
print("Precio final: $", precio_final)
````

### Pregunta 8: Condicional (1 punto)

Escribe un condicional que clasifique una `temperatura` segun estas reglas:

- Si `temperatura >= 30`: asigna `clima = "Caliente"`
- Si `temperatura >= 20` (y menor a 30): asigna `clima = "Templado"`
- Si `temperatura >= 10` (y menor a 20): asigna `clima = "Frio"`
- En otro caso: asigna `clima = "Muy frio"`

El codigo debe funcionar para cualquier valor de `temperatura`.

````python
# Pregunta 8: Clasifica la temperatura
temperatura = 30

if temperatura >= 30:
    clima = "Caliente"
elif temperatura >= 20:
    clima = "Templado"
elif temperatura >= 10:
    clima = "Frio"
else:
    clima = "Muy frio"

print("Temperatura:", temperatura, "- Clima:", clima)
````

### Pregunta 9: Bucle for (1 punto)

Usa un bucle `for` para calcular la suma de los numeros **pares** del 2 al 20 (incluido).

Guarda el resultado en la variable `suma_pares`.

````python
# Pregunta 9: Suma de pares del 2 al 20
suma_pares = 0

# Escribe tu bucle for aqui

for i in range(2, 20):
    if i % 2 == 0:
        suma_pares += i

print("Suma de pares del 2 al 20:", suma_pares)
````

### Pregunta 10: Bucle + condicional (1 punto)

Dada la siguiente lista de numeros, usa un bucle `for` para contar cuantos son mayores a 50.

Guarda el resultado en la variable `cuenta_mayores`.

```python
numeros = [12, 85, 37, 64, 50, 91, 23, 55, 8, 72]
```

````python
# Pregunta 10: Contar numeros mayores a 50
numeros = [12, 85, 37, 64, 50, 91, 23, 55, 8, 72]
cuenta_mayores = 0

# Escribe tu bucle aqui

for numero in numeros:
    if numero > 50:
        cuenta_mayores += 1

print("Numeros mayores a 50:", cuenta_mayores)
````
