```python
# Diccionario de productos y precios
productos = {
    "arroz": 3500,
    "frijoles": 4200,
    "aceite": 8500,
    "azucar": 3000,
    "sal": 1500
}

# Lista de compras (nombre del producto y cantidad)
compras = [
    ("arroz", 2),
    ("aceite", 1),
    ("azucar", 3)
]

# Calcula el total de la compra
total = 0
print("Detalle de compra:")
for producto, cantidad in compras:
    precio = productos[producto]
    subtotal = precio * cantidad
    total = total + subtotal
    print(f"{producto}: ${precio} x {cantidad} = ${subtotal}")

print(f"\nTotal a pagar: ${total}")

# Agrega un nuevo producto al diccionario
productos["cafe"] = 12000  
print("\nProductos actualizados:", productos)
```

Ejercicio 3.2

```python
# Funcion para buscar estudiante por nombre
def buscar_estudiante(lista_estudiantes, nombre_buscar):
    """
    Busca un estudiante en la lista por su nombre.
    Retorna el diccionario del estudiante si lo encuentra,
    None si no lo encuentra.
    """
    for estudiante in lista_estudiantes:
        if estudiante["nombre"] == nombre_buscar:
            return estudiante
    return None

# Prueba la funcion
nombre = "Ana Garcia"
resultado = buscar_estudiante(estudiantes, nombre)

if resultado != None:
    print(f"Estudiante encontrado: {resultado}")
else:
    print(f"Estudiante {nombre} no encontrado")

# Ahora busca un estudiante que no existe
nombre2 = "Alejandro Bermudez"
resultado2 = buscar_estudiante(estudiantes, nombre2)
if resultado2 != None:
    print(f"Estudiante encontrado: {resultado2}")
else:
    print(f"Estudiante {nombre2} no encontrado")
```

ejercicio 3.3

```python
# Funcion para calcular promedio de un estudiante
def calcular_promedio(notas):
    """
    Calcula el promedio de una lista de notas.
    """
    return sum(notas) / len(notas)

# Funcion para calcular promedio general de todos
def promedio_general(lista_estudiantes):
    """
    Calcula el promedio de todos los estudiantes.
    """
    total = 0
    cantidad_estudiantes = len(lista_estudiantes)

    for estudiante in lista_estudiantes:  
        promedio = calcular_promedio(estudiante["notas"])
        print(f"{estudiante['nombre']}: {promedio:.2f}")
        total = total + promedio

    return total / cantidad_estudiantes

# Calcula y muestra el promedio general
print("\nPromedios individuales:")
promedio_total = promedio_general(estudiantes)
print(f"\nPromedio general del grupo: {promedio_total:.2f}")

# Encuentra el estudiante con mejor promedio
mejor_estudiante = None
mejor_promedio = 0

for estudiante in estudiantes:
    promedio = calcular_promedio(estudiante["notas"])
    if promedio > mejor_promedio:
        mejor_promedio = promedio
        mejor_estudiante = estudiante["nombre"]

print(f"\nMejor estudiante: {mejor_estudiante}")
```
