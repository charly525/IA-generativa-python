### Pregunta 7: Funciones con return (1 punto)

Crea una funcion llamada `calcular_imc(peso, estatura)` que reciba el peso en kilogramos y la estatura en metros, y **retorne** el Indice de Masa Corporal (IMC).

Formula: `IMC = peso / estatura ** 2`

Luego, usa la funcion para calcular el IMC con `peso = 70` y `estatura = 1.75`, y guarda el resultado en `resultado_imc`.

```python
# Pregunta 7: Funcion calcular_imc

# Escribe tu funcion aqui
def calcular_imc(peso, estatura):
    imc = peso / (estatura ** 2)
    return imc

# Calcula el IMC con los valores dados
peso = 70
estatura = 1.75
resultado_imc = calcular_imc(peso, estatura) 


print("IMC:", resultado_imc)
```

### Pregunta 8: Funcion con valor por defecto (1 punto)

Crea una funcion llamada `saludar(nombre, idioma="espanol")` que:

- Si `idioma` es `"espanol"`, **retorne** `"Hola, {nombre}!"`
- En cualquier otro caso, **retorne** `"Hello, {nombre}!"`

Luego, guarda los resultados en las variables indicadas.

```python
# Pregunta 8: Funcion saludar con valor por defecto

# Escribe tu funcion aqui
def saludar(nombre, idioma="espanol"):
    if idioma == "espanol":
        return f"Hola, {nombre}!"
    elif idioma == "ingles":
        return f"Hello, {nombre}!"
    else:
        return f"Idioma no soportado para {nombre}."

# Prueba tu funcion
saludo_espanol = saludar("Ana")  
saludo_ingles = saludar("Ana", "ingles") 


print("Espanol:", saludo_espanol)
print("Ingles:", saludo_ingles)
```

### Pregunta 9: Listas (1 punto)

Dada la siguiente lista de numeros, crea:

1. Una lista llamada `mayores_20` que contenga solo los elementos **mayores a 20**
2. Una variable `suma_mayores` con la **suma** de los elementos de `mayores_20`

```python
# Pregunta 9: Filtrar lista y sumar
numeros = [15, 8, 42, 23, 4, 16, 33, 7, 19, 50]

# Escribe tu codigo aqui
mayores_20 = [num for num in numeros if num > 20] 
suma_mayores = sum(mayores_20)


print("Mayores a 20:", mayores_20)
print("Suma:", suma_mayores)
```

### Pregunta 10: Diccionarios (1 punto)

Dado el siguiente diccionario con las notas de los estudiantes, crea:

1. Un diccionario llamado `aprobados` que contenga solo los estudiantes con nota **>= 3.5**
2. Una variable `cantidad_aprobados` con el **numero** de estudiantes aprobados

```python
# Pregunta 10: Filtrar diccionario
estudiantes = {"Ana": 4.5, "Luis": 3.2, "Maria": 4.8, "Carlos": 2.9, "Sofia": 3.7}

# Escribe tu codigo aqui
aprobados = {nombre: nota for nombre, nota in estudiantes.items() if nota >= 3.5} 
cantidad_aprobados = len(aprobados)


print("Aprobados:", aprobados)
print("Cantidad:", cantidad_aprobados)
```
