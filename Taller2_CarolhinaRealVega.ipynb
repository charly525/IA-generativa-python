Ejercicio 2.2

```python
def calcular_descuento (monto):
    if 50000 <= monto < 100000:
        return monto - monto * 0.10
    elif monto >= 100000:
        return monto - monto * 0.20
    else:
        return monto
    pass

print(f"Monto final: ${calcular_descuento(30000)}")
print(f"Monto final: ${calcular_descuento(75000)}")
print(f"Monto final: ${calcular_descuento(150000)}")
```

Ejercicio 2.3

```python
def clasificar_edad(edad):
    if edad < 0:
        return "Edad no válida"
    elif edad <= 12:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 64:
        return "Adulto"
    else:
        return "Adulto mayor"
pass
    
print(clasificar_edad(8))
print(clasificar_edad(15))
print(clasificar_edad(35))
print(clasificar_edad(65))
```

Ejercicio 3.1

```python
def calcular_factorial(n):
    for i in range(1,n):
        n *= i
    return n
pass

print(f"5! = {calcular_factorial(5)}")
print(f"7! = {calcular_factorial(7)}")
```

ejercicio 3.2

```python
def encontrar_maximo(lista):
    for numero in lista:
        if numero == max(lista):
            return numero
pass

numeros = [23, 45, 12, 67, 34, 89, 15]
print(f"El maximo es: {encontrar_maximo(numeros)}")

numeros2 = [5, 2, 8, 1, 9, 3]
print(f"El maximo es: {encontrar_maximo(numeros2)}")
```

ejercicio 3.3

```python
def contar_vocales(texto):
    vocales = "aeiou"
    contador = 0
    for letra in texto.lower():
        if letra in vocales:
            contador += 1
    return contador
pass

frase1 = "Hola Mundo" 
print(f"El numero de vocales: {contar_vocales(frase1)}")
frase2 = "Python es un lenguaje de programacion"
print(f"El numero de vocales: {contar_vocales(frase2)}")
```
